```python
# from goatools.obo_parser import GODag

def get_level_n_go_terms(go,namespace="biological_process",level=1):
    """
    Get all level 1 GO terms under the specified namespace (e.g., biological process).
    
    Args:
        go_file (str): Path to the GO OBO file (e.g., go-basic.obo).
        namespace (str): Namespace of interest ('biological_process', 'molecular_function', 'cellular_component').
    
    Returns:
        dict: A dictionary where keys are GO IDs and values are term names.
    """
    # Load the GO DAG (Directed Acyclic Graph)
#     go_dag = GODag(go_file)
    
    # Find all level 1 terms under the specified namespace
    level_n_terms = {
        go_id: go_term.name
        for go_id, go_term in go.items()
        if go_term.level == level and go_term.namespace == namespace
    }
    
    return level_n_terms
# # Example usage
# go_file = "path/to/go-basic.obo"  # Download from http://current.geneontology.org/ontology/go-basic.obo
# level_1_terms = get_level_1_go_terms(go_file)

# # Print level 1 GO terms
# print("Level 1 GO Terms under Biological Process:")
# for go_id, name in level_1_terms.items():
#     print(f"{go_id}: {name}")
```

```python
import gzip

def get_genes_for_go_term_gz(gaf_gz_file, go_id, taxon_id="taxon:9606"):
    """
    Extract human gene symbols associated with a specific GO term ID from a compressed GAF file.
    
    Args:
        gaf_gz_file (str): Path to the compressed GAF file (e.g., goa_human.gaf.gz).
        go_id (str): GO term ID of interest (e.g., "GO:0008150").
        taxon_id (str): Taxonomy ID for humans (default is "taxon:9606").
    
    Returns:
        set: A set of unique gene symbols associated with the GO term.
    """
    genes = set()
    
    with gzip.open(gaf_gz_file, 'rt') as file:  # 'rt' opens the file in text mode
        for line in file:
            if line.startswith("!"):  # Skip comment lines
                continue
            
            columns = line.strip().split("\t")
            if len(columns) < 13:
                continue  # Skip malformed lines
            
            # Check for matching GO ID and taxon ID
            if columns[4] == go_id and taxon_id in columns[12]:
                genes.add(columns[2])  # Column 2: DB Object Symbol (gene symbol)
    
    return list(genes)
```

```python
go_union_names = [goid + '_' + goname.replace(" ", ".") for goid, goname in go_union.items()]
go_union_names[:5]
```

```python
# generate dictionary of GO-BP id to genes
go_union_db = {}
for i in go_union_names:
    go_union_db[i] = get_genes_for_go_term_gz('goa_human.gaf.gz',
                                              i.split('_')[0],
                                              taxon_id="taxon:9606")
```
